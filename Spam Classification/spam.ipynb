```python
import pandas as pd
import numpy as np
```

```python
data = pd.read_csv("spam.csv", encoding="latin1")
```

```python
data.head()
```

```python
from sklearn.preprocessing import LabelEncoder
le = LabelEncoder()
en = le.fit_transform(data["v1"])
data["v1"] = en
data["v1"].unique()
```

```python
data.shape
```

```python
#  data.drop(columns=["Unnamed: 2","Unnamed: 3","Unnamed: 4"], inplace=True)
```

STEP 1: TOKENIZATION

```python
import re
import nltk
from nltk.corpus import stopwords
from nltk.stem import WordNetLemmatizer
```

```python
sentences = []
v = data["v2"]
for i in range(len(v)):
    sent = nltk.sent_tokenize(v[i])
    sentences.append(sent)
```

```python
len(sentences)
```

```python
lemmatizer = WordNetLemmatizer()
```

```python
stop_words = set(stopwords.words("english"))
corpus = []
for sent in sentences:
    sent = " ".join(sent)
    review = re.sub('[^a-zA-Z0-9]',' ',sent)
    review = review.lower()
    review = review.split()
    review = [lemmatizer.lemmatize(word) for word in review if word not in stop_words] 
    review = ' '.join(review)
    corpus.append(review)
           
```

```python
len(sentences)
```

```python
len(corpus)
```

```python
corpus
```

### WORD TO VECTORS USING TF-IDF

```python
from sklearn.feature_extraction.text import TfidfVectorizer
tfidf = TfidfVectorizer(max_features=2500,ngram_range=(1,2))
Y = tfidf.fit_transform(corpus)
```

```python
tfidf.vocabulary_
```

```python
corpus[0]
```

```python
Y[0].toarray()
```

```python
y = Y.toarray()
y.shape
```

```python
d = pd.DataFrame(y)
d.shape
```

```python
d.head()
```

```python
data.head(10)
```

```python
d["output"] = data["v1"]
```

```python
d.head(10)
```

```python
d.shape
```

### *NOW READY TO BUILD ML MODEL* 

```python
x = d.iloc[:,:-1]
y = d["output"]
```

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test = train_test_split(x,y,test_size=0.20,random_state=42)
```

##### LOGISTIC REGRESSION

```python
from sklearn.linear_model import LogisticRegression
lr = LogisticRegression()
lr.fit(x_train,y_train)
```

```python
lr.score(x_train,y_train)*100
```

```python
lr.score(x_test,y_test)*100
```

```python
y_pred = lr.predict(x_test)
```

```python
from sklearn.metrics import mean_squared_error,accuracy_score,classification_report
acs = accuracy_score(y_pred,y_test)
mse = mean_squared_error(y_test,y_pred)
cf = classification_report(y_pred,y_test)
print("************ lOGISTIC REGRESSION ***************\n")
print("MSE: ",mse*100,"%")
print("Accuracy Score: ",acs*100,"%")
print("Confusion Matrix Report : \n",cf)
```
